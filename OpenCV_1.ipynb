## Reading Images with OpenCV

```python
#Import libraries
import cv2
import os
from matplotlib import pyplot as plt
```

```python
#Create image path
image_path = os.path.join('images', 'orange_kitty.jpg')

#Read image with cv2
image = cv2.imread(image_path)

#Display image with plt
plt.imshow(image)
```

```python
#Matplotlib uses BGR format. We need to convert to RGB format
recolor = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Display recolored image
plt.imshow(recolor)
plt.show()
```

```python
#Convert to Gray format
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Display gray image
plt.imshow(gray)
plt.show()
```

```python
#Save Image
image_out_path = os.path.join('images', 'grayimage.jpg')
cv2.imwrite(image_out_path, gray)
```
